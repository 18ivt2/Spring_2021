#### 8.	Решите поставленную задачу классификации в соответствии с заданием. При подборе параметров классификатора используйте метод GridSearchCV и перекрёстную проверку 

##### 8.1 Классификатор K ближайших соседей (задаётся количество ближайших объектов)

```python
Y = A[:,0]
X = A[:,1:]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=1000, random_state=5)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
classifier = KNeighborsClassifier()
print(classifier.get_params())
```

```python
parameters = {
    "p": [1, 2],
    "weights": ["uniform", "distance"],
    "n_neighbors": [1, 2, 3, 4, 5, 6, 7, 8, 9]
}
```

```python
gridSearch = GridSearchCV(classifier, parameters, n_jobs=4)
# Создаем GridSearchCV. Передаем наш классификатор и его параметры
gridSearch.fit(X_train, Y_train)
# Запуск GridSearchCV. Для каждого значения параметра из 'parameters' он запустит классификатор на тестовом датасете
# признаков (X_train) и сравнит ответ классификатора с настоящим (Y_train)
```

```python
print(gridSearch.best_params_)
```

```python
cvResults = pd.DataFrame(gridSearch.cv_results_)
cvResults[["param_p", "param_n_neighbors", "param_weights", "mean_test_score"]]
# полный отчет по параметрам
# наилучший результат мы получаем тогда, когда колличество ближайших объектор равно единице при равномерном весе 
# наихудший результат мы получаем при количестве объектов больше 7 при равномерном весе (его максимум равен 0,999)
# при весе точек, обратному расстоянию, значения столбца mean_test_score принимают значение, равное единице при любом весе
```

```python
# Выбранные параметры классификатора: p = 1, param_weights = 'uniform', n_neighbors=1
classifier_good = KNeighborsClassifier(n_neighbors=1, p=1, weights='uniform')
classifier_good.fit(X_train, Y_train)
Y_predicted_good = classifier_good.predict_proba(X_test)
```

```python
# PR-кривая (кривая между точностью и отзывов)
from sklearn.metrics import plot_precision_recall_curve, average_precision_score, recall_score
disp_good = plot_precision_recall_curve(classifier_good, X_test, Y_test)
disp_good.ax_.set_title('Precision-Recall curve (good)')
```

##### 8.2 Логический классификатор (бинарное решающее дерево).

```python
from sklearn.tree import DecisionTreeClassifier
treeClassifier = DecisionTreeClassifier()
```

```python
print(treeClassifier.get_params())
```

```python
parameters = {
    "max_depth": [3, 6, 9, 12, 15], # Максимальная глубина дерева. Если нет, то узлы расширяются до тех пор, пока все листья не станут чистыми 
    "max_features": [50, 90, 110], # Количество функций, которые следует учитывать при поиске лучшего разделения
    "min_samples_leaf": [3, 5, 10, 90] # Минимальное количество выборок, необходимых для нахождения в листовом узле.
}
```

```python
gridSearchTree = GridSearchCV(treeClassifier, parameters, n_jobs=4)
# Создаем GridSearchCV. Передаем наш классификатор и его параметры
gridSearchTree.fit(X_train, Y_train)
# Запуск GridSearchCV. Для каждого значения параметра из 'parameters' он запустит классификатор на тестовом датасете
# признаков (X_train) и сравнит ответ классификатора с настоящим (Y_train)
```

```python
print(gridSearchTree.best_params_)
```

```python
cvResults = pd.DataFrame(gridSearchTree.cv_results_)
cvResults[["param_max_depth", "param_max_features", "param_min_samples_leaf", "mean_test_score"]]
```

```python
# лучший выбор параметров: {'max_depth': 9, 'max_features': 50, 'min_samples_leaf': 3}
tree_classifier_good = DecisionTreeClassifier(max_depth=9, max_features=50, min_samples_leaf=3)
tree_classifier_good.fit(X_train, Y_train)
tree_Y_predicted_good = tree_classifier_good.predict_proba(X_test)

# Для сравнения создадим второй классификатор с худшими параметрами.
tree_classifier_bad = DecisionTreeClassifier(max_depth=3, max_features=50, min_samples_leaf=5)
tree_classifier_bad.fit(X_train, Y_train)
tree_Y_predicted_bad = tree_classifier_bad.predict_proba(X_test)
```

```python
from sklearn.metrics import plot_precision_recall_curve, average_precision_score, recall_score


disp_good = plot_precision_recall_curve(tree_classifier_good, X_test, Y_test)
disp_bad = plot_precision_recall_curve(tree_classifier_bad, X_test, Y_test)
disp_good.ax_.set_title('Precision-Recall curve (good)')
disp_bad.ax_.set_title('Precision-Recall curve (bad)')


average_precision_good = average_precision_score(Y_test, tree_Y_predicted_good[:,1])
average_precision_bad = average_precision_score(Y_test, tree_Y_predicted_bad[:,1])
recall_good = recall_score(Y_test, np.around(tree_Y_predicted_good[:,1]))
recall_bad = recall_score(Y_test, np.around(tree_Y_predicted_bad[:,1]))

print("Good average precision:", average_precision_good, "recall:", recall_good)
print("Bad average precision:", round(average_precision_bad, 2), "recall:", round(recall_bad, 2))
```

Задача бинарной классификации гриба на типы "съедобный"/"несъедобный" успешно решена с использованием двух типов классификаторов: KNeighborsClassifier и DecisionTreeClassifier.
При правильном подборе параметров каждого классификатора удалось добиться значений метрик precision и recall по 100%, что говорит об абсолютной точности работы классификатора на тестовой выборке. Также показано, что не оптимальный выбор параметров приводит к ухудшению работы классификатора (значения precision и recall уменьшаются).
