```python
import pandas as pd #Импортируем библиотеку pandas
mushrooms_df = pd.read_csv('mushrooms.csv') #Создаем dataframe из файла csv
mushrooms_df
```

```python
#Удалим дубликаты строк
mushrooms_df.drop_duplicates(keep=False)
```

```python
mushrooms_df.info() #Выводим информацию о dataframe'e
```

```python

import warnings
warnings.filterwarnings('ignore')
import os
import re
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
from sklearn.metrics import roc_auc_score, roc_curve, confusion_matrix
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.datasets import fetch_20newsgroups, load_files
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score

#Заменим категориальные признаки с помощью LabelEncoder
label_encoder = LabelEncoder()
categorial_columns = mushrooms_df.columns[mushrooms_df.dtypes == 'object'].union(['class'])
for column in categorial_columns:
    mushrooms_df[column] = label_encoder.fit_transform(mushrooms_df[column])
mushrooms_df.head()
```

```python
#Оценим сбалансированность выборки
mushrooms_df['class'].plot(kind = 'hist')
```

```python
#Проверим данные на пропуски
mushrooms_df.columns[mushrooms_df.isna().any()].tolist()
```

```python

X = mushrooms_df.iloc[:, 1:22] #Признаки
y = mushrooms_df.iloc[:, 0] #Классы

#Поделим датасет на тренировочную и тестовую выборку
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.2)
```

```python
#Определяем классификатор (RadiusNeighborsClassifier)
#Подберём наилучшие параметры для классификатора с помощью GridSearchSV
from sklearn.model_selection import GridSearchCV
#Определяем классификатор (RadiusNeighborsClassifier) не внося в него параметров
from sklearn.neighbors import RadiusNeighborsClassifier
classifier = RadiusNeighborsClassifier()
#Создадим словарь параметров, которые будем проверять в GridSearchSV для выбора лучшего
parameteres = {'radius': range(1, 20, 1),
              'leaf_size': range(10, 100, 10)}
grid = GridSearchCV(classifier, parameteres, cv = 5)
grid.fit(X_train, y_train)
```

```python
#Выведем лучшие параметры для нашей модели
grid.best_params_
```

```python
#Переопределим модель с лучшими параметрами
classifier = RadiusNeighborsClassifier(radius=3, leaf_size=10)
#Запустим обучение
classifier.fit(X_train, y_train)
```

```python
#Оценим качество работы классификатора
classifier.score(X_test, y_test)
```

```python
classifier.score(X_train, y_train)
```

```python
#Поделим датасет на тренировочную и тестовую выборку
X = mushrooms_df.iloc[:, 1:21]
y = mushrooms_df.iloc[:, 0]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.2)
```

```python
#Определяем классификатор (Perceptron)
from sklearn.linear_model import Perceptron
net = Perceptron()
parameteres = {'max_iter': range(10, 200, 5)}
grid = GridSearchCV(net, parameteres, cv = 5)
grid.fit(X_train, y_train)
```

```python
#Выведем лучшие параметры для нашей модели
grid.best_params_
```

```python
net = Perceptron(max_iter = 30, verbose = 0, random_state = None, fit_intercept = True, eta0 = 0.001,
                penalty = 'l2', alpha = 0.03)
net.fit(X_train, y_train)
```

```python
net.score(X_train, y_train)
```

```python
net.score(X_test, y_test)
```
