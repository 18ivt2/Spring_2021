## 2) Вычислите корреляцию Пирсона и Спирмена между признаками Age и G. Сравните полученные величины корреляции. Постройте гистограммы для указанных признаков. Оцените значение p-value.



```python
import numpy as np
import scipy
from scipy import stats

df = Seasons_Stats_df[['Age','G']].dropna()
age = np.asarray(df['Age'])
game = np.asarray(df['G'])
Pirson = scipy.stats.pearsonr(age, game)
Spirmen = scipy.stats.spearmanr(age, game)

print("Корреляция Пирсона:", Pirson)

if Pirson[1] < 0.05:
    print('p-value меньше 0.05, статистически незначима')
else:
    print('p-value больше 0.05, статистически значима ')
    
print("Корреляция Спирмена:", Spirmen)    
if Spirmen[1] < 0.05:
    print('p-value меньше 0.05, статистически незначима')
else:
    print('p-value больше 0.05, статистически значима')
    
Seasons_Stats_df.Age.hist(bins = 30)
Seasons_Stats_df.G.hist(bins = 30)
```
