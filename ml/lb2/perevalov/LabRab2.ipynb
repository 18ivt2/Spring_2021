### 1.Визуализируйте корреляции между признаками, находящимися в наборе данных (heatmap). Сделайте выводы.

### 2.Постройте график счастья по регионам, страны на графике должны представляться отдельными точками.

```python
df_all = df_2015.sort_values(by='Region', ascending=False)
df_all = df_all.reset_index(drop=True) # Сбрасываем индекс для нормальной нумерации

plt.ylabel('Happiness Score')
x = 0
region = df_all['Region'][0] # Перебор начнём с первого региона
_x = [0]
labels = [df_all['Region'][0]] # Название регионов

for i in range(0, df_2015.last_valid_index()+1): # Отмечаем количество счастья стран по регионам
    if region is not df_all['Region'][i]: # Если перешли к странам другого региона
        labels.append(df_all['Region'][i])
        region = df_all['Region'][i]
        x = x + 1
        _x.append(x)
    plt.plot(x, df_all['Happiness Score'][i], 'bo')
    
plt.xticks(_x, labels, rotation='vertical')
plt.show()
```

### 3.Оцените количество счастья по годам по регионам.

```python
pd.DataFrame({
    '2015 Score': df_2015.groupby(['Region'])['Happiness Score'].mean(), # Группируем по региону и находим
    '2016 Score': df_2016.groupby(['Region'])['Happiness Score'].mean()  # среднее значение кол-ва счастья
})
```

### 4.Определите наиболее сильно изменяющиеся параметры по разным странам в разные годы.

### 7. Вычислите корреляцию Пирсона и Спирмена между признаками Trust.Government.Corruption и Family. Сравните полученные величины корреляции. Постройте гистограммы для указанных признаков. Оцените значение p-value.

```python
import scipy.stats as stats

df_all = pd.concat([df_2015, df_2016])

# При вычислении корреляций передаются два параметра между, которыми её необходимо найти
# Возвращается пара значений: Сам критерий и p-значение
pearson_corr, pearson_pval = stats.pearsonr(df_all['Family'], df_all['Trust (Government Corruption)'])
spearman_corr, spearman_pval = stats.spearmanr(df_all['Family'], df_all['Trust (Government Corruption)'])

# Строим гистограмму
sns.histplot(df_all['Trust (Government Corruption)'].tolist())
sns.histplot(df_all['Family'].tolist())

print("Критерий Пирсона: {0}. P-значение: {1};".format(pearson_corr, pearson_pval))
print("Критерий Спирмена: {0}. P-значение: {1}".format(spearman_corr, spearman_pval))
```
