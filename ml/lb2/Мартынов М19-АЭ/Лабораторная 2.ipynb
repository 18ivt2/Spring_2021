### 1 Визуализируйте корреляции между признаками, находящимися в наборе данных (heatmap). Сделайте выводы

```python
correlation_heatmap = year2017.drop(['Country', 'Happiness.Rank', 'Happiness.Score'], axis=1).corr()
sns.heatmap(correlation_heatmap, annot=True, fmt='.if', linewidths = .5)
```

Из данной таблицы и ее визуализации можно сделать вывод, что поля: Whisker.high и Whisker.low коррелируют, т.е. взаимосвязаны друг с другом. ### 3 Оцените количество счастья по годам по регионам.

```python
year2015_1 = year2015.loc[:, ('Country', 'Region')]
year2016_1 = year2016.loc[:, ('Country', 'Region')]
year2015_2016 = year2015_1.merge(year2016_1, how='outer')
year2015_2016_2017 = year2017.merge(year2015_2016, on='Country')
pd.DataFrame({
    '2015': year2015.groupby(['Region'])['Happiness Score'].mean(),
    '2016': year2016.groupby(['Region'])['Happiness Score'].mean(),
    '2017': year2015_2016_2017.groupby(['Region'])['Happiness.Score'].mean(),
})
```

### 4 Определите наиболее сильно изменяющиеся параметры по разным странам в разные годы.

```python
year2015 = year2015.assign(year=pd.Series(2015, index=year2015.index))
year2016 = year2016.assign(year=pd.Series(2016, index=year2016.index))
year2017 = year2017.assign(year=pd.Series(2017, index=year2017.index))
df_all = pd.concat([year2015, year2016, year2017])
df_all = pd.DataFrame(df_all.sort_values(by='Country', ascending=True))
df_all = df_all.reset_index(drop=True)
print(df_all[['Dystopia Residual', 'Economy (GDP per Capita)', 'Family', 'Freedom', 'Generosity',
             'Health (Life Expectancy)', 'Lower Confidence Interval', 'Standard Error',
             'Trust (Government Corruption)', 'Upper Confidence Interval']].idxmax(axis=1) + ' / ' + df_all['Country']
     + ' / ' + df_all['year'].apply(str)) 
```

### 7 Вычислите корреляцию Пирсона и Спирмена между признаками Trust.Government.Corruption и Family. Сравните полученные величины корреляции. Постройте гистограммы для указания признаков. Оцените значение p-value.

```python
print(stats.pearsonr(year2015['Trust (Government Corruption)'], year2015['Family']))
print(stats.spearmanr(year2015['Trust (Government Corruption)'], year2015['Family']))
sns.distplot(year2015['Trust (Government Corruption)'])
if stats.spearmanr(year2015['Trust (Government Corruption)'], year2015['Family'])[1] < 0.05:
    print('Приемлимое значение p-value для корреляции Спирмена')
else:
    print('Не приемлимое значение p-value для корреляции Спирмена')
if stats.pearsonr(year2015['Trust (Government Corruption)'], year2015['Family'])[1] < 0.05:
    print('Приемлимое значение p-value для корреляции Пирсона')
else:
    print('Не приемлимое значение p-value для корреляции Пирсона')
```
