```python
#9 Вычислите корреляцию Пирсона и Спирмена между признаками Gender и Missions. 
#Сравните полученные величины корреляции. 
#Постройте гистограммы для указанных признаков. Оцените значение p-value.
```

```python
from sklearn.preprocessing import LabelEncoder
import scipy
import scipy.stats
import pandas as pd 
import numpy as np

label_encoder = LabelEncoder()


gender = np.asarray(astronauts['Space Walks'])
missions = np.asarray(label_encoder.fit_transform (astronauts['Missions']))


pirson = scipy.stats.pearsonr(gender, missions)
spirmen = scipy.stats.spearmanr(gender, missions)

print('пиросон = ', pirson)

if (pirson[1] < 0.01):
    print("p-value < 0.01 незначителен")
else:
    print("p-value > 0.01 - значителен")

print("Спирмен", spirmen)

if (spirmen[1] < 0.01):
    print("p-value < 0.01 незначителен")
else:
    print("p-value > 0.01 - значителен")
```
