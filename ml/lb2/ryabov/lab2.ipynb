# Задание 1
## Определить астронавта, который провел наибольшее время в открытом космосе. Построить график, на котором по оси ОХ отображены астронавты, а по оси  OY время проведенное в открытом космосе.

```python
astronauts_df.sort_values(by=['Space Walks (hr)'], ascending = False).head(1)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = astronauts_df['Name']
y = astronauts_df['Space Walks (hr)']
plt.figure(figsize=(20, 10))
plt.plot(x,y)
plt.xlabel('Astronaut(index)',fontsize = 20)
plt.ylabel('Time(hr)',fontsize = 20)
plt.xticks(np.arange(0,357,1),rotation=90)
plt.yticks(np.arange(0,70,5),fontsize = 20)
plt.title('Time in open space',fontsize = 20)
```

# Задание 2
## Укажите уничерситет, выпустивший наибольшее количество астронавтов. Постройте график, на котором по оси ОХ отображены университеты, а по оси OY количество астронавтов, которые учились в данном университете.

```python
university_df = pd.DataFrame(astronauts_df.groupby(['Alma Mater'])['Name'].count())
```

```python
university_df.sort_values(by=['Name'], ascending = False).head(1)
```

```python
x = university_df.index
y = university_df['Name']
plt.figure(figsize=(20, 10))
plt.plot(x,y)
plt.xlabel('Name of university',fontsize = 20)
plt.ylabel('Number of astronauts',fontsize = 20)
plt.xticks(rotation=90)
plt.yticks(np.arange(0,13,1),fontsize = 20)
plt.title('Universities that have produced the largest number of astronauts',fontsize = 20)
```

# Задание 9
## Вычислите корреляцию Пирсона и Спирмена между признаками  Gender и Space Walks (hr). Сравните полученные результаты. Постройте гистограммы для указанных признаков. Оцените значение p-value.

```python
from scipy.stats import pearsonr
from scipy.stats import spearmanr
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
label_encoder.fit(astronauts_df.Gender)
astronauts_df['Gender']=label_encoder.transform(astronauts_df.Gender)

astronauts_df['Gender'].corr(astronauts_df['Space Walks (hr)'])


```

```python
astronauts_df['Gender'].corr(astronauts_df['Space Walks (hr)'], method = 'spearman')
```
