```python
# 1.Визуализируйте корреляции между признаками, находящимися в наборе данных (heatmap). Сделайте выводы.
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df_2015 = pd.read_csv(r"C:\Win\WPy64-3870\notebooks\Lab №2\2015.csv")
df_2015 = df_2015.drop(['Standard Error'], axis='columns')
df_2015
```

```python
df_2016 = pd.read_csv(r"C:\Win\WPy64-3870\notebooks\Lab №2\2016.csv")
df_2016 = df_2016.drop(['Lower Confidence Interval'], axis='columns')
df_2016 = df_2016.drop(['Upper Confidence Interval'], axis='columns')
df_2016
```

```python
# 7.Вычислите корреляцию Пирсона и Спирмена между признаками Trust.Government.Corruption и Family.
# Сравните полученные величины корреляции. Постройте гистограммы для указанных признаков. Оцените значение p-value.
df_2015_cor = df_2015[['Country', 'Trust (Government Corruption)', 'Family']]
df_2015_cor
```

```python
from scipy.stats.stats import pearsonr
pearsonr(df_2015_cor['Trust (Government Corruption)'], df_2015_cor['Family'])
```

```python
df_2015_cor.corr(method='pearson')
```

```python
from scipy.stats.stats import spearmanr
spearmanr(df_2015_cor['Trust (Government Corruption)'], df_2015_cor['Family'])
```

```python
df_2015_cor.corr(method='spearman')
```
