```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
ir = pd.read_csv("rainfall in india 1901-2015.csv").rename(columns=str.lower)
dr = pd.read_csv("district wise rainfall normal.csv").rename(columns=str.lower)
```

# Задание 1: Постройте графики количества осадков по годам в разных штатах.

```python
ir_graph = ir.groupby(['subdivision'])
fig,ax = plt.subplots(figsize=(8,6))

for label, ir_new in ir_graph:
    ir_new.plot(kind='line', ax=ax, label=label, x='year', y='annual', figsize=(20,10))
plt.show()
```

# Задание 2: Постройте графики количества осадков по месяцам по всем годам кумулятивно.

```python
ir_cum = ir[['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec']]
ir_cum = ir_cum.cumsum()
ir_cum.plot(figsize=(10,10))
plt.show()
```

# Задание 3: Постройте графики количества осадков по штатам, используйте boxplot.

```python
fig = plt.figure(figsize=(20, 10))
plt.xticks(rotation='vertical')
sns.boxplot(x='subdivision', y='annual', data=ir)
```
