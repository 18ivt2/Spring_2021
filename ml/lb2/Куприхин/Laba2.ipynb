## 4. Определите время года, в которое вероятность появления метеоритов максимальна. Постройте график.

```python

def get_month(month):
    return month.split('-')[1].strip()
def get_season(season):
    if (season == '03') or (season == '04') or (season == '05'):
        season = 'spring'
    if (season == '06') or (season == '07') or (season == '08'):
        season = 'summer'
    if (season == '09') or (season == '10') or (season == '11'):
        season = 'autumn'
    if (season == '12') or (season == '01') or (season == '02'):
        season = 'winter'
    return season
cneos_fireball['Season'] = cneos_fireball['Peak Brightness Date/Time (UT)'].apply(get_month)
cneos_fireball['Season'] = cneos_fireball['Season'].apply(get_season)
cneos_fireball.groupby(['Season'])['Season'].count().plot(kind = "bar")
```

```python
print(cneos_fireball['Season'].value_counts().index[0],'=',cneos_fireball['Season'].value_counts().max())
```

## 6.Построить график, на котором по оси ОХ отложено время суток, по OY – частота появления метеоритов.

```python
def get_hour (date):
    time = date.split(' ')[1].strip()
    return time.split(':')[0].strip()

cneos_fireball['Hour'] = cneos_fireball['Peak Brightness Date/Time (UT)'].apply(get_hour)
cneos_fireball.groupby(['Hour'])['Hour'].count().plot()
```
