# Задание 1

### Постройте графики количества осадков по годам в разных штатах. 

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
sub1 = pd.DataFrame(rains_df[rains_df['SUBDIVISION'] == 'TAMIL NADU'])
```

```python
x=sub1['YEAR']
y=sub1['ANNUAL']
plt.figure(figsize = (15,7))
plt.xticks(np.arange(1901,2016,2),rotation=90)
plt.plot(x,y)
plt.xlabel('Год')
plt.ylabel('Количество осадков')
plt.title('Количество осадков по годам в штате TAMIL NADU')
```

```python
sub2 = pd.DataFrame(rains_df[rains_df['SUBDIVISION'] == 'WEST RAJASTHAN'])
```

```python
x=sub2['YEAR']
y=sub2['ANNUAL']
plt.figure(figsize = (15,7))
plt.xticks(np.arange(1901,2016,2),rotation=90)
plt.plot(x,y)
plt.xlabel('Год')
plt.ylabel('Количество осадков')
plt.title('Количество осадков по годам в штате WEST RAJASTHAN')
```

# Задание 9

#### Вычислите корреляцию Пирсона и Спирмена между признаками NOV и MAR. 

#### Корреляция Пирсона

```python
rains_df['NOV'].corr(rains_df['MAR'])      
```

#### Корреляция Спирмена

```python
rains_df['NOV'].corr(rains_df['MAR'], method='spearman')
```

#### Вывод: Корреляция Спирмена больше корреляции Пирсона. Коэффициент Пирсона: p-value меньше 0.05, значит статистически незначима. Коэффициент Спирмена: p-value больше 0.03, значит связь между признаками очень слабая.


```python
rains_df['NOV'].hist(bins=30)
```
