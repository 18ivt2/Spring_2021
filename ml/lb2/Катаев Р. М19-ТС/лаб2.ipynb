```python
import pandas as pd #импортируем библиотеку pandas 
cameras_df = pd.read_csv('camera_dataset.csv') #создаем dataframe из csv-файла датасета 
cameras_df.head() #выводим первые 5 строчек в dataframe
```

```python
cameras_df.info() #выводим информацию dataframe
```

```python
cameras_df = cameras_df.drop(index=[0]) #удаляем строку с индексом 0
cameras_df.head() 
```

```python
cameras_df = cameras_df.reset_index(drop=True) #обновили индексы
cameras_df.head() 
```

```python
cameras_df = cameras_df.astype({'Model':'string','Release date': 'double', 'Max resolution' : 'double', 'Low resolution' : 'double', 'Effective pixels' : 'double',
                  'Zoom wide (W)' : 'double', 'Zoom tele (T)' : 'double', 'Normal focus range' : 'double', 
                   'Macro focus range' : 'double', 'Storage included':'double', 'Weight (inc. batteries)':'double',
                  'Dimensions':'double', 'Price':'double'}) 
#изменяем типы данных признаков
```

```python
cameras_df.dtypes #выводим типы данных
```

```python
#Удаляем строки с нулями в значениях
import numpy as np
cameras_df = cameras_df.replace(0, np.nan)
cameras_df = cameras_df.dropna(how='any', axis=0)
cameras_df.head()
```

```python
#Обновим индексы и упорядочим по годам
cameras_df = cameras_df.sort_values('Release date', ascending=True).reset_index(drop=True).astype({'Model':'string'})
cameras_df.head()
```

```python
#Проверим размер фрейма
cameras_df.info()
```

```python
#Выберем столбцы с ценой и годом
price_df  = cameras_df[['Release date','Price']]
price_df.head()
```

```python
#Построим график средней цены от года
#Подготовим данные
mean_price = price_df.groupby(['Release date']).mean()
#mean_price
mean_price = mean_price.reset_index()
mean_price
```

```python
#График изменения средней цены 
import matplotlib
matplotlib.style.use('ggplot')

mean_price.set_index('Release date')['Price'].plot()
```

```python
#Наиболее часто встречающаяся компания в наборе данных
#Сначала выделим компанию из строк в Model
import re
cameras_df['Company'] = cameras_df['Model'].apply(lambda x : re.findall(r"[\w']+",x)[0])
cameras_df
```

```python
#Выведем количество строк по уникальным названиям компаний. value_counts() выводит количество строк в порядке убывания.
cameras_df['Company'].value_counts() 
#Ответ: Olympus
```

```python
#Добавим в набор данных признак, являющийся произведением двух других признаков.
cameras_df['New column'] = cameras_df['Low resolution'] *cameras_df['Effective pixels']
cameras_df.head()
```

```python
#Определим камеру с максимальной и минимальной стоимостью.

#Подготовим данные
max_price = cameras_df[['Model','Price']]
max_price.sort_values('Price', ascending=False).reset_index(drop=True)
#Самая дорогая - Ricoh RDC-i500
#Самая дешевая - JVC GC-QX5HD
```

```python
#Определите год, в который было выпущено максимально количество новых камер.
cameras_df['Release date'].value_counts()
#Ответ - в 2007.
```

```python
import scipy.stats

#Коэф. корреляции Пирсона и p-value
x = cameras_df['Price']
y = cameras_df['Low resolution']

print(scipy.stats.pearsonr(x, y))

x = cameras_df['Price']
y = cameras_df['Max resolution']

print(scipy.stats.pearsonr(x, y))

#Коэф. корреляции Спирмена и p-value
x = cameras_df['Price']
y = cameras_df['Low resolution']

print(scipy.stats.spearmanr(x, y))

x = cameras_df['Price']
y = cameras_df['Max resolution']

print(scipy.stats.spearmanr(x, y))
```
